```python
import matplotlib.pyplot as plt
y=[1,3,6,8,9,15,17,20]
x1=[1,3,6,8,9,11,13,14]
x2=[1,7,9,10,17,20,25,28]
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
line1=ax.plot(x1,y,'gs-')
line2=ax.plot(x2,y,'co--')
ax.legend(labels=('movies', 'webseries'),loc='lower right')
ax.set_title("users preferences")
ax.set_xlabel('medium')
ax.set_ylabel=('preferance')
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(10)
 
data1 = np.random.normal(100, 10, 200)
data2 = np.random.normal(90, 20, 200)
data3 = np.random.normal(80, 30, 200)
data4 = np.random.normal(70, 40, 200)
data = [data1, data2, data3, data4]
 
fig = plt.figure(figsize =(10, 7))
 
ax = fig.add_axes([0, 0, 1, 1])
 
bp = ax.boxplot(data)
ax.set_yticklabels(['data1','data2','data3','data4'])
plt.title("box plot")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=['A','B','C','D']
y1=[10,20,30,20]
y2=[20,30,40,10]
y3=[20,5,30,10]
y4=[20,30,40,15]
plt.bar(x,y1,color='c')
plt.bar(x,y2,bottom=y1,color='y')
plt.bar(x,y3,bottom=y2,color='b')
plt.bar(x,y4,bottom=y3,color='g')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend(["R1","R2","R3","R4"])
plt.title("scores")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.axes(projection='polar')
r=3
rads=np.arrange(0,(2*np.pi),0.1)
for rad in rads:
    plt.polar(rad,r,'g.')
plt.legend(label='polar chart',loc=1)
plt.title("POLAR CHART")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
  
feature_x = np.linspace(-5.0, 3.0, 70)
feature_y = np.linspace(-5.0, 3.0, 70)
  
[X, Y] = np.meshgrid(feature_x, feature_y)
  
fig, ax = plt.subplots(1, 1)

Z = X ** 2 + Y ** 2
  
cp=ax.contourf(X, Y, Z)
fig.colorbar(cp)
  
ax.set_title('Filled Contour Plot')
ax.set_xlabel('x axis')
ax.set_ylabel('y axis')
  
plt.show()
```
