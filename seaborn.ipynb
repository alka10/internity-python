```python
import matplotlib.pyplot as plt 
import pandas as pd 
import numpy as np 
import seaborn as sns
from warnings import filterwarnings
df=pd.read_csv('placement_data_full_class.csv')
data=df['degree_p']
sns.distplot(data)
plt.show()
```

```python
sns.jointplot(x='degree_p',y='degree_t',data=df)
plt.show()
```

```python
sns.jointplot(x='ssc_p',y='hsc_p',data=df,kind='kde')
plt.show()
```

```python
sns.pairplot(df,palette='coolwarm')
plt.show()
```

```python
sns.rugplot(df['salary'])
plt.show()
```

```python
sns.set_style('darkgrid')
sns.barplot(x='gender',y='salary',data=df,palette='plasma')
plt.show()
```

```python
sns.countplot(x='gender',data=df)
plt.show()
```

```python
sns.boxplot(x='gender',y='salary',data=df)
plt.show()
```

```python
sns.stripplot(x ='gender',y='salary',data=df,jitter=True,dodge=True)
plt.show()
```

```python
sns.violinplot(x='gender',y='salary',data=df)
plt.show()
```

```python
sns.set_style('whitegrid')
sns.lmplot(x='salary',y='degree_p',data=df)
plt.show()
```

```python
sns.lmplot(x ='salary', y ='degree_p', data = df,
          markers =['o'])
plt.show()
```
